```python
import random
secret_number=random.randint(1,10)
guess_count=0
max_guess=5
print("Guess a number from 1 to 10")
print("You have maximum of 5 guesses")
while guess_count<max_guess:
    num1=int(input("Enter your number"))
    guess_count+=1
    if num1<secret_number:
        print("The secret number is higher")
    elif num1>secret_number:
        print("The secret number is lower")
    else:
        print(f"Helll Yeahhh!! You guessed in {guess_count} attempts")
        break
else:
    print(f"Better luck next time!!The secret number was {secret_number}")
```
